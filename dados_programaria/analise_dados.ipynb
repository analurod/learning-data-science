## 4.2 -Tratando valores faltantes (parte 1)

### Principais Formas de Tratar
- **Excluir dados nulos**: A vantagem é que é rápido e simples, porém não indicados. A desvantagem é que perdemos dados e em caso de poucos dados isso é de grande significância.
- **Substituir pela média ou mediana**: A vantagem é que não perdemos nenhum dado. Funciona bem em bases pequenas ou com dados homogêneos. Com dados categoricos: podemos utilizar a moda. A desvantagem é não levar em conta a covariância dos fatores (valores que variam em conjunto).
- **Modelo de Aprndizagem de Máquina**: A vantagem é que leva em conta covariância dos dados. A desvantagem é que o metódo é mais trabalhoso qu os demais.

```python
# Verificando Na's
dados.info()
```

A tabela tem 4271 linhas com dados, porém temos algumas váriavéis com valores menores de linhas não-nulas, como por exemplo:
- GÊNERO: 4262
- IDADE: 4197
- SALÁRIO: 3694

### Trabalhando coluna de Gênero

```python
# Contagem linhas por gênero
dados.groupby('GENERO', dropna=False)['ID'].nunique()
```

Há 12 "Prefiro Não Informar" e 9 nulos. Neste caso, como havia a opção de marcar "Prefiro Não Informar" podemos preencher os valores nulos dessa forma.

```python
# Preencher com "Prefiro não informar"
dados['GENERO'] = dados['GENERO'].fillna("Prefiro não informar")

# Verificar
dados.groupby('GENERO', dropna=False)['ID'].nunique()
```

## 4.2 -Tratando valores faltantes (parte 2)

### Trabalhando coluna de idade

```python
# Verificando nulos quantidade de nulos
dados["IDADE"].isnull().value_counts()
```

Nesse caso, temos 74 valores nulos, o que representa aproximadamente 1,74% dos dados. Dessa forma, não compensaria o trabalho de criar um modelo para tratar os valores nulos. Vamos utilizar a média das idades nos valores nulos.

```python
# Verificar colunas - Procurar informações relevantes para auxiliar no preenchimento
dados.columns
```

Aqui temos a coluna "Faixa Idade". Pode ser útil, será que as pessoas que não preencheram a idade informaram a faixa de idade?

```python
# Verificar quantas pessoas não preencheram idade mas preencheram faixa etária
dados[dados['IDADE'].isnull()]['FAIXA IDADE'].value_counts()
```

Todas as pessoas que deixaram a idade sem preenchimento, preencheram a faixa etária. Utilizarei essa informação para preencher os nulos de forma mais assertiva.

```python
# Média 17-21
media_17_21 = dados[dados['FAIXA IDADE']=="17-21"]["IDADE"].mean()
```

```python
# Localizar linhas e colunas onde idade vazia e faixa etária 17-21
dados.loc[(dados['FAIXA IDADE']=="17-21") & (dados['IDADE'].isnull()), 'IDADE'] = media_17_21
```

```python
# Verificar
dados[dados['IDADE'].isnull()]['FAIXA IDADE'].value_counts()
```

```python
#  Média 55+
dados[dados['FAIXA IDADE']=="55+"]["IDADE"].mean()
```

Média vazia, nenhuma pessoa 55+ preencheu a idade. Talvez seja um erro de formulário, necessário perguntar e entender, mas nesse caso usaremos média geral.

```python
#  Média Geral
media_geral = dados["IDADE"].mean()
```

```python
# Localizar linhas e colunas onde idade vazia e faixa etária 55+
dados.loc[(dados['FAIXA IDADE']=="55+") & (dados['IDADE'].isnull()), 'IDADE'] = media_geral
```

```python
# Verificar
dados[dados['IDADE'].isnull()]['FAIXA IDADE'].value_counts()
```

São 577 linhas nulas, vamos verificar se há faixa salarial
