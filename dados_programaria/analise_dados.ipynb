### Principais Formas de Tratar
- **Excluir dados nulos**: A vantagem é que é rápido e simples, porém não indicados. A desvantagem é que perdemos dados e em caso de poucos dados isso é de grande significância.
- **Substituir pela média ou mediana**: A vantagem é que não perdemos nenhum dado. Funciona bem em bases pequenas ou com dados homogêneos. Com dados categoricos: podemos utilizar a moda. A desvantagem é não levar em conta a covariância dos fatores (valores que variam em conjunto).
- **Modelo de Aprndizagem de Máquina**: A vantagem é que leva em conta covariância dos dados. A desvantagem é que o metódo é mais trabalhoso qu os demais.

## 4.4 - Intervalo de confiança e distribuição amostral

 - **Média Amostral**: medida de tendência central calculada a partir de um subconjunto de dados (amostra).

- **Desvio Padrão Amostral**: O desvio padrão nos mostra o quanto os valores variam em relação à média amostral.

- **Erro Padrão**: O erro padrão é calculado dividindo o desvio padrão da amostra pela raiz quadrada do tamanho da amostra. Ele nos ajuda a entender a precisão da nossa média amostral como uma estimativa da média populacional.

- **Intervalo de Confiança de 95%**: Calculamos o intervalo de confiança para um nível de 95% de confiança. Isso significa que, se repetíssemos esse processo muitas vezes, 95% das vezes o intervalo conteria a verdadeira média populacional.


**Importância do Intervalo de Confiança**
**Precisão:** O intervalo de confiança nos dá uma faixa dentro da qual acreditamos que a média populacional se encontra, tornando nossa estimativa mais informativa do que uma única média amostral.

**Incerteza:** Reconhece a incerteza inerente à amostragem e nos dá uma ideia de quanto essa incerteza pode ser.

```python
# Pegar todos os salários
salario = dados['SALARIO']
```

```python
salario
```

```python
# Média salários
media_amostral = np.mean(salario)
media_amostral
```

```python
# Desvio Padrão salários
dp_amostral = np.std(salario)
dp_amostral
```

```python
nivel_confianca = 0.95
```

```python
tamanho_amostra = len(salario)
tamanho_amostra
```

```python
# Erro Padrão
erro_padrao = stats.sem(salario) # dividindo o desvio padrão da amostra pela raiz quadrada do tamanho da amostra
erro_padrao
```

```python
# Intervalo de confiança com nivel de confiança de 95%
ic = stats.t.interval(nivel_confianca, tamanho_amostra-1, loc=media_amostral, scale= erro_padrao)
ic
```

## 4.6 - Correlação, diferentes funções para dados discretos e contínuos

A correlação nos diz se e como duas variáveis estão relacionadas. Existem dois tipos principais de correlação: positiva e negativa.

- **Correlação positiva**: ocorre quando duas variáveis aumentam juntas. Por exemplo, quanto mais um aluno estuda, maior é seu desempenho acadêmico.

- **Correlação negativa**: ocorre quando uma variável aumenta enquanto a outra diminui. Por exemplo, quanto mais tempo um aluno passa assistindo TV, menor é seu desempenho acadêmico.

A função de correlação mais usada é a de Pearson, que varia de -1 a 1. Um valor próximo de 1 indica uma forte correlação positiva, próximo de -1 indica uma forte correlação negativa, e próximo de 0 indica que não há uma relação linear.

Por exemplo, ao calcular a correlação entre horas de estudo e desempenho acadêmico, um coeficiente de 0.8 indica uma correlação positiva forte. 

- Correlação entre variáveis discretas: coeficiente de Cramér. Zero indica nenhuma associação e Um associação Completa.

```python
# Correlação entre idade e salario
correlacao_continua = dados['IDADE'].corr(dados['SALARIO'])
correlacao_continua
```

Correlação positiva, porém fraca entre idade e salário.
Para identificar correlações em tabelas com muitas colunas de valores contínuos, utilizamos mapas de calor. 

```python
# Correlação de cramer
def cramer_coeficiente(coluna1, coluna2):
    # Tabela Cruzada -> Frequencia de que os dados das tabelas se cruzam
    tabela_cruzada = np.array(pd.crosstab(coluna1, coluna2)) 

    chi2 = chi2_contingency(tabela_cruzada)[0]

    soma = np.sum(tabela_cruzada)

    mini = min(tabela_cruzada.shape)-1

    cramer = np.sqrt(chi2/(soma*mini))

    return cramer
```

```python
cramer_coeficiente(dados['COR/RACA/ETNIA'], dados['NIVEL DE ENSINO'])
```

```python
cramer_coeficiente(dados['GENERO'], dados['NIVEL DE ENSINO'])
```
